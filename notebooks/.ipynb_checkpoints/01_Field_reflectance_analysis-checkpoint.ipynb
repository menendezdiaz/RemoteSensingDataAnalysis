# 01 - Field reflectance analysis

This notebook loads, clean and visualices field reflectance data from different types of soil and vegetation. 

The data was obtained using the FieldSpec® 3 portable spectroradiometer. This instrument integrates three sensors, enabling measurements across a wide spectral range—from the visible to the near-infrared regions. It features precision optics with 25° and 10° field-of-view options and a fiber optic cable for collecting spectral data under different conditions. The device is equipped with a backpack for field use, a handheld probe for precise targeting, and dedicated software for instrument control and data analysis. Measurements are referenced against a Spectralon panel, which serves as a calibration target with near-100% reflectance.

Spectroradiometer sensor specifications:

| Sensor | Wavelength Range (nm) | Spectral Resolution (nm) | Sampling Interval (nm) |
|---------|------------------------|---------------------------|------------------------|
| NIR     | 400–1000               | 3                         | 1.4                    |
| SWIR1   | 1000–1800              | 10                        | 2                      |
| SWIR2   | 1800–2400              | 10                        | 2                      |



The objective of this exercise is to measure the spectral signature of different materials in the visible and near-infrared regions.

```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path


# ------------------------------------------------------
# 1️⃣ CONFIGURACIÓN INICIAL
# ------------------------------------------------------

# Carpeta de datos (relativa al notebook)
data_folder = Path('../data/field')

# Diccionario de clases: cada categoría con sus objetos
categorias = {
    'superficies': ['black_asphalt', 'blue_paint', 'wood_table', 'concrete'],
    'suelos': ['soil', 'dry_grass', 'rough_soil'],
    'vegetacion': ['lemon_leaf', 'red_flower', 'rosemary', 'jatropha', 'small_flower', 'weed', 'grass']
}

# Franjas malas (en nm)
def apply_mask(wavelength, reflectance):
    mask = (
        ((wavelength >= 1350) & (wavelength <= 1451)) |
        ((wavelength >= 1810) & (wavelength <= 1970)) |
        (wavelength >= 2320)
    )
    reflectance[mask] = np.nan
    reflectance[reflectance > 1] = np.nan
    reflectance[reflectance < 0] = np.nan
    return reflectance

```

```python

# ------------------------------------------------------
# 2️⃣ CARGA Y PROMEDIO DE DATOS
# ------------------------------------------------------

def cargar_firmas_por_objeto(folder, nombre_base):
    """
    Carga todos los archivos de un objeto (ej: asphalt1.txt, asphalt2.txt, ...)
    y devuelve la longitud de onda y la media de reflectancia.
    """
    archivos = sorted(folder.glob(f'{nombre_base}*.txt'))
    if not archivos:
        print(f'⚠️ No se encontraron archivos para {nombre_base}')
        return None, None
    
    firmas = []
    for archivo in archivos:
        data = np.loadtxt(archivo, skiprows=1)
        wl, refl = data[:,0], data[:,1]
        refl = apply_mask(wl, refl)
        firmas.append(refl)
    
    # Promedio ignorando NaN
    mean_refl = np.nanmean(firmas, axis=0)
    return wl, mean_refl
```

```python

# ------------------------------------------------------
# 3️⃣ REPRESENTACIÓN DE CADA CATEGORÍA
# ------------------------------------------------------

def plot_categoria(categoria, lista_objetos):
    plt.figure(figsize=(12,6))
    
    for obj in lista_objetos:
        wl, mean_refl = cargar_firmas_por_objeto(data_folder, obj)
        if wl is not None:
            plt.plot(wl, mean_refl, label=obj)
    
    plt.title(f'Firmas espectrales — {categoria.capitalize()}')
    plt.xlabel('Wavelength [nm]')
    plt.ylabel('Reflectance')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

# ------------------------------------------------------
# 4️⃣ GENERAR LAS 3 GRÁFICAS PRINCIPALES
# ------------------------------------------------------

for categoria, objetos in categorias.items():
    plot_categoria(categoria, objetos)
```
